```python
hotel_name1=[]
city1=[]
Location1=[]
no_of_reviews1=[]
user_rating1=[]
star_ratings1=[]
review_score1=[]
```

```python
url_hotel=[]

for i in link:
    driver.get(i)
    url_tags=driver.find_elements(By.XPATH,'//header[@class="bui-spacer--medium"]//a')
    for j in url_tags:
        href=j.get_attribute('href')
        url_hotel.append(href)
    
```

```python
df=pd.DataFrame(columns=['S_no','Hotel_name','Cities','Location','No_of_reviews','User_ratings','Star_Rating','Review_score'])
for k in url_hotel:
    
    driver.get(k)
    hotel_name=driver.find_element(By.XPATH,'//h2[@class="d2fee87262 pp-header__title"]').text
    city=driver.find_element(By.XPATH,'//input[//@class="eb46370fe1"]').get_attribute('value')
    Location=driver.find_element(By.XPATH,'//*[@id="showMap2"]/span[1]').text
    no_of_reviews=driver.find_element(By.XPATH,'//div[@class="abf093bdfe f45d8e4c32 d935416c47"]').text
    no_of_reviews=no_of_reviews.split(' ')[0]
    no_of_reviews=no_of_reviews.replace(',' , '')
    star_ratings=driver.find_element(By.XPATH,'//div[@class="a3b8729ab1 d86cee9b25"]').text
    user_rating=driver.find_element(By.XPATH,'//div[@class="best-review-score best-review-score-with_best_ugc_highlight hp_lightbox_score_block"]//span').text
    review_score=driver.find_element(By.XPATH,'//div[@class="a3b8729ab1 e6208ee469 cb2cbb3ccb"]').text
    
    hotel_name1.append(hotel_name)
    city1.append(city)
    Location1.append(Location)
    no_of_reviews1.append(no_of_reviews)
    user_rating1.append(user_rating)
    star_ratings1.append(star_ratings)
    review_score1.append(review_score)
    #print('hotel_name1')
    
df['Hotel_name']=hotel_name1
df['Cities']=city1
df['Location']=Location1
df['No_of_reviews']=no_of_reviews1
df['User_ratings']=user_rating1
df['Star_Rating']=star_ratings1 
df['Review_score']=review_score1
df['S_no']=np.arange(1,len(df)+1)
df      
       
```

```python
data1=df.to_csv('Hotel_info_1.csv',index=True)
data1
```

```python

df1=pd.DataFrame(columns=['S_no','Hotel_name','Restaurant','Bar','Heating','Lift','Breakfast_Cuisine1','Breakfast_Cuisine2','Breakfast_Cuisine3'])

facilities=['Free WiFi','Family rooms','Non-smoking rooms','Lift','Heating','Restaurant','Bar']

hotel_name=[]
Breakfast_Cuisine1_value=[]
Breakfast_Cuisine2_value=[]
Breakfast_Cuisine3_value=[]

n=0


for link in url_hotel:
    driver.get(link)
    hotel_name_value=driver.find_element(By.XPATH,'//h2[@class="d2fee87262 pp-header__title"]').text
    hotel_name.append(hotel_name_value)
    


    for j in facilities:
        found='no'
        #print('searching for ',j)
        #print('-------------------------')
        div_tags=driver.find_elements(By.XPATH,'//div[@class="a53cbfa6de ebbf62ced0 a623bc7ac6"]')
        for k in div_tags:
           
            if(k.text==j):
                found='yes'
                df1.at[n,j]='Yes'  
                
        item_tags= driver.find_elements(By.XPATH,'//span[@class="a5a5a75131"]')
        for i in item_tags:
            #print(i.text)
            if(i.text == j):
                found='yes'
               # print(found)
               # print(j,'found')
                df1.at[n,j]='Yes'
                
               
        if(found=='no'):
            df1.at[n,j]="No"
    n=n+1   
    
   #For Breakfast cuisine

    try:
        bf_cuisine=driver.find_element(By.XPATH,'//span[@class="ph-item-copy-breakfast-option"]').text
        bf_cuisine_items=bf_cuisine.split(',')
        if(len(bf_cuisine_items)==1):

            Breakfast_Cuisine1_value.append(bf_cuisine_items[0])
            Breakfast_Cuisine2_value.append(np.nan)
            Breakfast_Cuisine3_value.append(np.nan)
        elif(len(bf_cuisine_items)==2):
            Breakfast_Cuisine1_value.append(bf_cuisine_items[0])
            Breakfast_Cuisine2_value.append(bf_cuisine_items[1])
            Breakfast_Cuisine3_value.append(np.nan)
        else: 
            Breakfast_Cuisine1_value.append(bf_cuisine_items[0])
            Breakfast_Cuisine2_value.append(bf_cuisine_items[1])
            Breakfast_Cuisine3_value.append(bf_cuisine_items[2])
            
    except:
        
        Breakfast_Cuisine1_value.append('No')
        Breakfast_Cuisine2_value.append('No')
        Breakfast_Cuisine3_value.append('No')
        
df1['S_no']=np.arange(1,len(df1)+1)        
df1['Hotel_name']=hotel_name 
df1['Breakfast_Cuisine1']=Breakfast_Cuisine1_value
df1['Breakfast_Cuisine2']=Breakfast_Cuisine2_value
df1['Breakfast_Cuisine3']=Breakfast_Cuisine3_value
df1.fillna('No',axis=1,inplace=True)
df1          
```

```python

df1.rename(columns={'Free WiFi':'Free_wifi','Non smoking rooms':'Non_smoking_rooms','Family rooms':'Family_rooms'},inplace=True)
df1
```
