```python
#SQL Queries

# 1

query='SELECT * FROM HOTEL_INFO WHERE Cities="Sydney" and User_ratings>5'

result=cursor.execute(query)

for i in result:
    print(i)
```

```python
# 2

query='SELECT Cities,Location FROM HOTEL_INFO WHERE No_of_reviews=(select max(No_of_reviews) from hotel_info)'

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
# 3

query='SELECT * FROM HOTEL_INFO ORDER BY User_ratings DESC'

result=cursor.execute(query)

for row in result:
    print(row)
    
```

```python
#4

query='SELECT * FROM  HOTEL_INFO WHERE Star_rating>=4 ORDER BY Review_score DESC'


result=cursor.execute(query)

for row in result:
    print(row)
```

```python
#5

query='SELECT COUNT(Hotel_name),Cities FROM HOTEL_INFO GROUP BY Cities'

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
# Table 2

# 1
query='SELECT Hotel_name FROM HOTEL_FACILITIES WHERE Restaurant="Yes" AND Bar="Yes" '

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
# 2

query='SELECT Hotel_name FROM HOTEL_FACILITIES WHERE  Family_rooms="Yes" AND Free_wifi="Yes" '

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
# 3


query='SELECT DISTINCT(Breakfast_Cuisine1),COUNT(Breakfast_Cuisine1) FROM HOTEL_FACILITIES GROUP BY Breakfast_Cuisine1'
result=cursor.execute(query)

for row in result:
    print(row)
```

```python
# 4

query='SELECT Hotel_name FROM  HOTEL_FACILITIES WHERE Breakfast_Cuisine1="Continental" OR Breakfast_Cuisine2="Continental" OR Breakfast_Cuisine3="Continental" '

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
# 5

query='SELECT * FROM HOTEL_FACILITIES WHERE Heating="Yes" AND Lift="Yes" AND Bar="Yes" '

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
# Table 3

#1

query='SELECT DISTINCT(Hotel_name) FROM HOTEL_ROOM_TYPE WHERE Double_bed="Yes" ORDER BY Price DESC '

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
#2

query='SELECT Room_type,avg(price)  FROM HOTEL_ROOM_TYPE  GROUP BY Room_type'
count=1
result=cursor.execute(query)

for row in result:
    print(row)
    count=count+1
    
```

```python
result=cursor.execute('PRAGMA table_info("HOTEL_FACILITIES")')

for row in result:
    print(row)
```

```python

# Sub Queries
#3

query='SELECT * FROM HOTEL_INFO WHERE User_ratings>5 AND Hotel_name=(SELECT Hotel_name FROM HOTEL_FACILITIES WHERE Free_wifi="Yes") '

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
#4

query='SELECT DISTINCT(HOTEL_FACILITIES.HOTEL_NAME) FROM HOTEL_INFO,HOTEL_FACILITIES WHERE Family_rooms="Yes" AND Non_smoking_rooms="Yes" '

result=cursor.execute(query)

for row in result:
    print(row)
```

```python
#5

query='SELECT Room_type,AVG(price)  FROM HOTEL_ROOM_TYPE,HOTEL_FACILITIES  GROUP BY Room_type'
count=1
result=cursor.execute(query)

for row in result:
    print(row)
    count=count+1


```
